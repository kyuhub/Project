```python
#사전 형태로 만들기
id2token = dict(enumerate(words))
```

```python
#한글 불용어 사전을 추가한 extract_keywords
def extract_keywords(text):
    result = kiwi.analyze(text)
    for token, pos, start, end in result[0][0]:
        if (pos != 'NNB' and pos[0] in 'N') and (token not in korean_cafe_stopwords):
            yield f'{token}'
```

```python
docs = []
for text in cafe.review:
    doc = list(extract_keywords(text))
    docs.append(doc)
#cafe에 있는 모든 review를 토큰화를 해서 doc에 저장.
```

```python
docs
```

```python
from gensim.corpora.dictionary import Dictionary
```

```python
dic = Dictionary(docs)
```

```python
dic.filter_extremes(no_below=10, no_above = 0.9)
#filter_extremes : 극단적인 단어들을 기준으로 필터
#filter_n_most_frequent : 가장 자주 나오는 단어를 기준으로 필터
#no_below : 일부 문서에서만 사용하는 단어들을 제외(너무 안나오는 단어들을 제외)
#no_above : 너무 흔하게 나오는 단어들을 제외
#no_below=10, no_above = 0.9 : 10개 이하의 문서에서 나오는 단어들과 90% 이상의 문서에서 나오는 단어들은 제외
```

```python
len(dic)
```

## 토큰형식의 docs를 corpus 형식으로 바꾸기

```python
corpus = []
for doc in docs:
    bow = dic.doc2bow(doc) #bag of words : 단어들이 몇개씩 나왔는지 세서 묶어둔다.
    corpus.append(bow)
```

```python
corpus
```

```python
len(corpus)
```

## Corpus를 LDA 모델에 넣어서 주제분석 준비하기

```python
from gensim.models.ldamodel import LdaModel
```

```python
from sklearn.model_selection import train_test_split
train_corpus, valid_corpus = train_test_split(corpus, test_size=0.1, random_state=0)
```

```python
#LDA 모델을 돌릴 때 종종 LDA모델에서 DeprecationWarning 경고가 뜨기 때문에 경고를 끄자.
import warnings
warnings.filterwarnings('ignore', category = DeprecationWarning)
```

```python
import numpy
```

## 혼란도, 응집도, 다양도를 고려해서 topic의 수 구하기

```python
perplexity_list = []
coherence_list = []
diversity_list = []


for numtopic in range(1,101):
    model = LdaModel(corpus = train_corpus, id2word=dic, num_topics=numtopic, random_state=0)
    
    #혼란도 계산
    loss = model.log_perplexity(valid_corpus)
    print('numtopic', numtopic,'개의 혼란도: ', loss)
    perplexity_list.append(loss)
    
    #업데이트 혼란도(아래의 응집도와 다양도에 영향을 주기위해서)
    old_loss = -numpy.inf #음의 무한대
    
    while loss > old_loss + 0.1:
        model.update(train_corpus) #모델을 한번 더 추정해서 조금 더 정확한 추정치를 얻는다
        old_loss=loss
        loss = model.log_perplexity(valid_corpus)
    
    #응집도 계산
    coh = CoherenceModel(model=model, corpus=corpus, texts=docs, dictionary=dic, coherence='c_v')
    coherence = coh.get_coherence()
    print('numtopic', numtopic,'개의 응집도: ', coherence)
    coherence_list.append(coherence)
    
    #다양도 계산
    topn = 25
    top_words = set() #집합은 기존에 있는 원소면 추가가 안됨. -> 중복된 원소는 추가가 안됨.
    for topic in range(model.num_topics):
        for word, prob in model.show_topic(topic, topn=topn):
            top_words.add(word)
    diversity = len(top_words) / (numtopic*25)
    print('numtopic', numtopic,'개의 다양도: ', diversity)
    diversity_list.append(diversity)
    
    print('-----------------------------------------------------------------')
```

```python
len(perplexity_list), len(coherence_list), len(diversity_list)
```

```python
import pandas as pd
topic_selector = pd.DataFrame({'perplexity': perplexity_list, 'coherence': coherence_list, 'diversity': diversity_list})
```

```python
topic_selector.index = topic_selector.index + 1
topic_selector
```
