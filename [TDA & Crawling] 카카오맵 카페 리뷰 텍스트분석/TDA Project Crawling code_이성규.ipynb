```python
import os
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys
import time
```

## Crawling function

```python
def crawling(place):
    for i in place:
        #browser2로 각 카페 리뷰 사이트에 접속하기
        url = i.get_attribute('href')
        time.sleep(1)
        
        browser2.get(url)
        time.sleep(1)
        
        go = i.text[0]
        
        #카페 상세사이트의 1페이지 리뷰와 별점 수집하기 - 평가 개수가 0이 아닐 때평가 개수가 0이 아닐 때
        if int(go) != 0:
        
            #별점이 없는 '카카오플레이스 평가하기'의 수집은 제외하기 위해서 카카오맵 리뷰만 로딩
            try:
                browser2.find_element_by_css_selector('#mArticle > div.cont_evaluation > div.evaluation_sorting > a').send_keys(Keys.ENTER)
                browser2.find_element_by_css_selector('#mArticle > div.cont_evaluation > div.evaluation_sorting.open_sorting > ul > li:nth-child(2) > a').send_keys(Keys.ENTER)
                time.sleep(1)
            except:
                continue


            #카페 이름 수집하기
            ready_name = browser2.find_element_by_css_selector('div.inner_place h2.tit_location').text
            time.sleep(3)

  
            [review.append(r.text) for r in browser2.find_elements_by_css_selector('div.comment_info p.txt_comment')]
            [score.append(s.text) for s in browser2.find_elements_by_css_selector('div.grade_star.size_s em.num_rate')]
            [name.append(ready_name) for n in range(len(browser2.find_elements_by_css_selector('div.comment_info p.txt_comment')))]




            ##카페 상세사이트의 2~6페이지 리뷰와 별점 수집하기
            while True:
                idx=3
                try:
                    #다음버튼 전까지의 숫자페이지 리뷰와 별점 수집하기
                    page_num = len(browser2.find_elements_by_class_name('link_page'))
                    for i in range(page_num-1):
                        browser2.find_element_by_css_selector('#mArticle > div.cont_evaluation > div.evaluation_review > div > a:nth-child(' + str(idx) +')').send_keys(Keys.ENTER)
                        #2페이지가 (#mArticle > div.cont_evaluation > div.evaluation_review > div > a:nth-child(3))
                        time.sleep(1)

                        [review.append(r.text) for r in browser2.find_elements_by_css_selector('div.comment_info p.txt_comment')]
                        [score.append(s.text) for s in browser2.find_elements_by_css_selector('div.grade_star.size_s em.num_rate')]
                        [name.append(ready_name) for n in range(len(browser2.find_elements_by_css_selector('div.comment_info p.txt_comment')))]

                        idx = idx +1

                    #다음버튼을 누른 후 다음페이지 리뷰와 별점 수집하기
                    browser2.find_element_by_link_text('다음').send_keys(Keys.ENTER)
                    time.sleep(1)

                    [review.append(r.text) for r in browser2.find_elements_by_css_selector('div.comment_info p.txt_comment')]
                    [score.append(s.text) for s in browser2.find_elements_by_css_selector('div.grade_star.size_s em.num_rate')]
                    [name.append(ready_name) for n in range(len(browser2.find_elements_by_css_selector('div.comment_info p.txt_comment')))]

                except:
                    break

            #6~끝 페이지 리뷰와 별점 수집하기
            while True:
                idx=4
                try:
                    #다음버튼 전까지의 숫자페이지 리뷰와 별점 수집하기
                    page_num = len(browser2.find_elements_by_class_name('link_page'))
                    for i in range(page_num-1):
                        browser2.find_element_by_css_selector('#mArticle > div.cont_evaluation > div.evaluation_review > div > a:nth-child(' + str(idx) +')').send_keys(Keys.ENTER)
                        time.sleep(1)

                        [review.append(r.text) for r in browser2.find_elements_by_css_selector('div.comment_info p.txt_comment')]
                        [score.append(s.text) for s in browser2.find_elements_by_css_selector('div.grade_star.size_s em.num_rate')]
                        [name.append(ready_name) for n in range(len(browser2.find_elements_by_css_selector('div.comment_info p.txt_comment')))]

                        idx = idx +1

                    #다음버튼을 누른 후 다음페이지 리뷰와 별점 수집하기
                    browser2.find_element_by_link_text('다음').send_keys(Keys.ENTER)
                    time.sleep(1)

                    [review.append(r.text) for r in browser2.find_elements_by_css_selector('div.comment_info p.txt_comment')]
                    [score.append(s.text) for s in browser2.find_elements_by_css_selector('div.grade_star.size_s em.num_rate')]
                    [name.append(ready_name) for n in range(len(browser2.find_elements_by_css_selector('div.comment_info p.txt_comment')))]
                except:
                    print(browser2.find_element_by_css_selector('div.inner_place h2.tit_location').text, "데이터 수집 완료.") 
                    break
        else:
            continue
```

## Completed Crawling code
- '강릉 카페'

```python
#셀레늄 연결
browser = Chrome()
browser2 = Chrome()
time.sleep(1)

#browser1에 카카오맵 연결하기
browser.get('https://map.kakao.com/')
time.sleep(1)

#검색창에 검색어 입력
search_area = browser.find_element_by_css_selector('#search\.keyword\.query')
search_area.send_keys('강릉 카페')
time.sleep(1)

#검색하기 클릭
browser.find_element_by_css_selector('.go.ico_search.btn_search').send_keys(Keys.ENTER)
time.sleep(1)

#리스트 생성
name = []
review = []
score = []

#browser를 통한 1페이지 크롤링
cafes = browser.find_elements_by_css_selector('a.numberofscore')
time.sleep(1)
crawling(cafes)

#우선 1페이지의 더보기를 클릭
browser.find_element_by_css_selector('#info\.search\.place\.more').send_keys(Keys.ENTER)
time.sleep(1)

#2페이지~마지막페이지
while True:
    idx=2    
    try:
        #다음버튼 전까지의 페이지 크롤링 진행 후 페이지넘기기
        page_num = len(browser.find_elements_by_css_selector('div.pages'))
        for i in range(page_num-3):
            browser.find_element_by_css_selector('#info\.search\.page\.no'+ str(idx) +'').send_keys(Keys.ENTER)
            time.sleep(1)
            cafes = browser.find_elements_by_css_selector('a.numberofscore')
            time.sleep(1)
            crawling(cafes)
            idx = idx +1

        #다음버튼을 누른 후 해당 페이지 크롤링 진행 후 페이지넘기기
        browser.find_element_by_css_selector('#info\.search\.page\.next').send_keys(Keys.ENTER)
        time.sleep(1)
        cafes = browser.find_elements_by_css_selector('a.numberofscore')
        time.sleep(1)
        crawling(cafes)
    except:
        print('크롤링 끝!')
        break
```
